```python
import os
import pandas as pd
```

```python
file_path = 'WineDataset.csv'
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

df = df.dropna(subset=['Title', 'Description', 'Price','Capacity', 'Grape', 'Closure','Country', 'Unit', 'Characteristics', 'Per bottle / case / each', 'Type', 'ABV', 'Region', 'Style', 'Vintage'])
df
```

```python
# get all the unique characteristics in the dataset, separated by commas
unique_characteristics = set()
for row in df['Characteristics']:
    characteristics = row.split(',')
    for characteristic in characteristics:
        unique_characteristics.add(characteristic.strip())
(print(len(unique_characteristics)))
print(unique_characteristics)
```

```python
# print all descriptions to a csv file
df.to_csv('WineDataset_cleaned.csv', index=False)
```
