```python
import os
import pandas as pd
```

```python
file_path = 'wine_data.csv'
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)


df = df.dropna(subset=['Title', 'Description','Country', 'Region', 'Characteristics'])
df.head(5)
```

```python
# extract price from the Price 
# format of price is now in "£11.99 per botte"
df['Price'] = df['Price'].str.extract(r'£(\d+\.\d+)').astype(float)
```
